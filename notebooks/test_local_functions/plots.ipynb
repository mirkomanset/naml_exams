# Try the plot functions

```python
def quadratic_bowl(x, y):
    return x**2 + y**2

def rosenbrock(x, y):
    return (1 - x)**2 + 100 * (y - x**2)**2

def himmelblau(x, y):
    return (x**2 + y - 11)**2 + (x + y**2 - 7)**2
```

## Surface

```python
plot_surface(quadratic_bowl, -5, 5, 100)
```

```python
plot_surface(rosenbrock, -5, 5, 100)
```

```python
plot_surface(himmelblau, -5, 5, 100)
```
